```python
# 1. Import all the libraries we need
from sklearn import datasets                     # To load the iris dataset
from sklearn.model_selection import train_test_split # To split the data
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis # <-- This is the new model
from sklearn.metrics import classification_report      # For precision and recall
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay # For visualization
import matplotlib.pyplot as plt                # To show the plot

print("Libraries imported successfully.")
```

```python
# 2. Load and Prepare the dataset
# We already inspected it in the first notebook, so we know it has
# 4 predictors and 3 classes.
iris = datasets.load_iris()
X = iris.data  # Predictors
y = iris.target  # Classes (the 'answer')

print("Dataset loaded and prepared.")
```

```python
# 3. Split the Data
print("\n--- 1. Splitting the Data ---")
# The assignment says to use the *same* split for all methods.
# So, I MUST use the same random_state=42 and test_size=0.25
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Just printing to double-check the sizes are the same as before
print(f"Training set size: {len(X_train)}")
print(f"Testing set size: {len(X_test)}")


print("\n--- 2. Linear Discriminant Analysis (LDA) Model ---")
```

```python
# 4. Create and Train the Model
# Create an instance of the LDA model
lda_model = LinearDiscriminantAnalysis()

# Now, 'fit' (train) the model using the training data
lda_model.fit(X_train, y_train)
print("Model trained successfully.")
```

```python
# 5. Make Predictions
# Use the trained LDA model to predict answers for the test set
y_pred = lda_model.predict(X_test)


print("\n--- 3. Performance Report (Precision and Recall) ---")
```

```python
# 6. Evaluate the Model's Performance
# The assignment asks for precision and recall 
# We use classification_report again, comparing the true 'y_test' 
# with our new 'y_pred' from the LDA model.
report = classification_report(y_test, y_pred, target_names=iris.target_names)

print(report)

# We can see the weighted avg for precision and recall are both 1.00.
# This model also performed perfectly on this test set.


print("\n--- 4. Visualization ---")
```

```python
# 7. Visualize the Performance
# We'll make a confusion matrix for the LDA model
# as required by the assignment 

# Create the confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=lda_model.classes_)

# Use ConfusionMatrixDisplay to plot it
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)

print("Generating confusion matrix plot for LDA...")
# Show the plot
disp.plot()
plt.title("Confusion Matrix for Linear Discriminant Analysis")
plt.show()

print("\nLDA analysis complete.")
```
