```python
# 1. Import all the libraries we need
from sklearn import datasets                     # To load the iris dataset
from sklearn.model_selection import train_test_split # To split the data
from sklearn.neighbors import KNeighborsClassifier   # <-- This is the new model
from sklearn.metrics import classification_report      # For precision and recall
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay # For visualization
import matplotlib.pyplot as plt                # To show the plot

print("Libraries imported successfully.")
```

```python
# 2. Load and Prepare the dataset
# Load the iris dataset
iris = datasets.load_iris()
X = iris.data  # Predictors
y = iris.target  # Classes (the 'answer')

print("Dataset loaded and prepared.")
```

```python
# 3. Split the Data
print("\n--- 1. Splitting the Data ---")
# I MUST use the same random_state=42 and test_size=0.25
# This ensures my comparison between models is fair, as the assignment requires.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Printing to check the sizes are the same
print(f"Training set size: {len(X_train)}")
print(f"Testing set size: {len(X_test)}")


print("\n--- 2. K-Nearest Neighbour (KNN) Model ---")
```

```python
# 4. Create and Train the Model
# The assignment *specifically* asks for K=5
# In scikit-learn, this parameter is called 'n_neighbors'
k_value = 5
knn_model = KNeighborsClassifier(n_neighbors=k_value)
print(f"KNN model created with K={k_value}.")

# Now, 'fit' (train) the model using the training data
knn_model.fit(X_train, y_train)
print("Model trained successfully.")
```

```python
# 5. Make Predictions
# Use the trained KNN model to predict answers for the test set
y_pred = knn_model.predict(X_test)


print("\n--- 3. Performance Report (Precision and Recall) ---")
```

```python
# 6. Evaluate the Model's Performance
# The assignment asks for precision and recall
# We use classification_report again, comparing the true 'y_test' 
# with our new 'y_pred' from the KNN model.
report = classification_report(y_test, y_pred, target_names=iris.target_names)

print(report)

# Again, the weighted avg for precision and recall are both 1.00.
# This model also performed perfectly on this specific test set.


print("\n--- 4. Visualization ---")
```

```python
# 7. Visualize the Performance
# We'll make a confusion matrix for the KNN model
# as required by the assignment

# Create the confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=knn_model.classes_)

# Use ConfusionMatrixDisplay to plot it
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)

print("Generating confusion matrix plot for KNN...")
# Show the plot
disp.plot()
plt.title(f"Confusion Matrix for KNN (K={k_value})")
plt.show()

print("\nKNN analysis complete.")
```
